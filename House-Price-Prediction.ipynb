## Importing Required Dependencies 

```python
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

## Importing and Loading Dataset

```python
data=pd.DataFrame(pd.read_csv(r"C:\Users\dhyey\Downloads\HousingData.csv"))
```

```python
data["LSTAT"].fillna(data["LSTAT"].median(),inplace=True)
```

```python
data["CRIM"].fillna(data["LSTAT"].median(),inplace=True)
data["ZN"].fillna(data["LSTAT"].median(),inplace=True)
data["INDUS"].fillna(data["LSTAT"].median(),inplace=True)
data["CHAS"].fillna(data["LSTAT"].median(),inplace=True)
data["AGE"].fillna(data["LSTAT"].median(),inplace=True)

data.isnull().sum()
```

```python
x=data.drop(columns=['ZN','CHAS'],axis=1)
y=data[["MEDV"]]
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
r2_score(y_test,y_pred)
```

```python

model_rf = RandomForestRegressor()
model_rf.fit(x_train, y_train.values.ravel())  # Flatten y_train to 1D array

y_pred_rf = model_rf.predict(x_test)

mse_rf = mean_squared_error(y_test.values.ravel(), y_pred_rf)  # Flatten y_test for comparison
r2_rf = r2_score(y_test.values.ravel(), y_pred_rf)

print("Random Forest MSE:", mse_rf)
print("Random Forest R-squared:", r2_rf)
```

```python
y_train_pred = model_rf.predict(x_train)
r2_score(y_train,y_train_pred)
```
